# Cleaning the Data

```python
# to check if there are null values in dataset
print(df.isnull().sum())
```

```python
# by filling null values with 0
df = df.fillna(0)
# Again checking null values
print(df.isnull().sum())
```

```python
print(df.duplicated())        # To check for dupliacte values
```

# Data Visualization

```python
print(df.ssc_b.value_counts())
```

```python
figure, ax = plt.subplots(figsize=(8,6))
sns.countplot(x = 'ssc_b' , data =df, hue = 'status')
plt.show()
```

```python
figure, ax = plt.subplots(figsize=(8,6))
sns.countplot(x = 'specialisation' , data =df, hue = 'status')
plt.show()
```

```python
figure, ax = plt.subplots(figsize=(12,15))
sns.countplot(y = 'salary' , hue = 'gender', data =df)
plt.show()
```

```python
sns.histplot(df['salary'], bins=5)
```

# Relationship Analysis

```python
correlation = df.corr()
```

```python
# To check correlation using heatmap
figure, ax = plt.subplots(figsize=(10,8))
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot= True)
plt.show()
```

```python
#     To see each variable in out dataset
sns.pairplot(df)
```

```python
sns.relplot(x = 'salary', y = 'degree_p' , hue = 'gender', data = df)
```

```python
sns.catplot(x = 'degree_p', kind = 'box', data = df)
```

```python
sns.catplot(x="status", y="degree_p", data=df ,kind="swarm" ,hue='gender')
```

```python
# To check stream of students that are placed mostly
figure, ax = plt.subplots(figsize=(9,10))
sns.violinplot(x="degree_t", y="salary", data=df, palette = 'dark')
sns.stripplot(x="degree_t", y="salary", data=df, hue='status')
plt.show()
```
